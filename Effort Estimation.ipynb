```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt


#- Precedentedness (PREC)
#- Process Flexibility (FLEX)
#- Risk Resolution (RESL)
#- Team Cohesion (TEAM)
#- Process Maturity (PMAT)


# Generate universe variables
#   * According to COCOMO II factors
#   * https://numpy.org/doc/stable/reference/generated/numpy.arange.html
#x_PREC = np.arange(6.2 ,0.00 ,-1.24)
#x_FLEX = np.arange(5.07,0.00 ,-1.01 )
#x_RESL = np.arange(7.07,0.00 ,-1.41)
#x_TEAM = np.arange(5.48,0.00 ,-1.10)
#x_PMAT = np.arange(7.80,0.00 ,-1.56)

x_PREC = np.arange(0.00,7 ,1.24)
x_FLEX = np.arange(0.00,6 ,1.01 )
x_RESL = np.arange(0.00,8 ,1.41)
x_TEAM = np.arange(0.00,6 ,1.10)
x_PMAT = np.arange(0.00,8 ,1.56)

print ('x_PREC: ',x_PREC)
print ('x_FLEX: ',x_FLEX)
print ('x_RESL: ',x_RESL)
print ('x_TEAM: ',x_TEAM)
print ('x_PMAT: ',x_PMAT)


# Generate fuzzy membership functions for Precedentedness (PREC)
#    * https://pythonhosted.org/scikit-fuzzy/api/skfuzzy.html#trimf
PREC_VL = fuzz.trimf(x_PREC, [4.96 ,6.2 ,6.2])
PREC_L  = fuzz.trimf(x_PREC, [3.72,4.96,6.2])
PREC_N  = fuzz.trimf(x_PREC, [2.48,3.72,4.96])
PREC_H  = fuzz.trimf(x_PREC, [1.24,2.48,3.72])
PREC_VH = fuzz.trimf(x_PREC, [0,1.24,2.48])
PREC_XH = fuzz.trimf(x_PREC, [0,0,1.24])


# Generate fuzzy membership functions for Process Flexibility (FLEX)
#    * https://pythonhosted.org/scikit-fuzzy/api/skfuzzy.html#trimf
FLEX_VL = fuzz.trimf(x_FLEX, [4.05,5.07 ,5.07])
FLEX_L  = fuzz.trimf(x_FLEX, [3.04,4.05,5.07])
FLEX_N  = fuzz.trimf(x_FLEX, [2.03,3.04,4.05])
FLEX_H  = fuzz.trimf(x_FLEX, [1.01,2.03,3.04])
FLEX_VH = fuzz.trimf(x_FLEX, [0.00,1.01,2.03])
FLEX_XH = fuzz.trimf(x_FLEX, [0.00,0.00,1.01])


# Generate fuzzy membership functions for Risk Resolution (RESL)
#    * https://pythonhosted.org/scikit-fuzzy/api/skfuzzy.html#trimf
RESL_VL = fuzz.trimf(x_RESL, [5.65,7.07,7.07])
RESL_L  = fuzz.trimf(x_RESL, [4.24,5.65,7.07])
RESL_N  = fuzz.trimf(x_RESL, [2.83,4.24,5.65])
RESL_H  = fuzz.trimf(x_RESL, [1.41,2.83,4.24])
RESL_VH = fuzz.trimf(x_RESL, [0.00,1.41,2.83])
RESL_XH = fuzz.trimf(x_RESL, [0.00,0.00,1.41])

# Generate fuzzy membership functions for Team Cohesion (TEAM)
#    * https://pythonhosted.org/scikit-fuzzy/api/skfuzzy.html#trimf
TEAM_VL = fuzz.trimf(x_TEAM, [4.38,5.48,5.48])
TEAM_L  = fuzz.trimf(x_TEAM, [3.29,4.38,5.48])
TEAM_N  = fuzz.trimf(x_TEAM, [2.19,3.29,4.38])
TEAM_H  = fuzz.trimf(x_TEAM, [1.10,2.19,3.29])
TEAM_VH = fuzz.trimf(x_TEAM, [0.00,1.10,2.19])
TEAM_XH = fuzz.trimf(x_TEAM, [0.00,0.00,1.10])

# Generate fuzzy membership functions for Process Maturity (PMAT)
#    * https://pythonhosted.org/scikit-fuzzy/api/skfuzzy.html#trimf
PMAT_VL = fuzz.trimf(x_PMAT, [6.24,7.80,7.80])
PMAT_L  = fuzz.trimf(x_PMAT, [4.68,6.24,7.80])
PMAT_N  = fuzz.trimf(x_PMAT, [3.12,4.68,6.24])
PMAT_H  = fuzz.trimf(x_PMAT, [1.56,3.12,4.68])
PMAT_VH = fuzz.trimf(x_PMAT, [0.00,1.56,3.12])
PMAT_XH = fuzz.trimf(x_PMAT, [0.00,0.00,1.56])

# Visualize these universes and membership functions
fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(nrows=5, figsize=(10, 15))

# Visualize Precedentedness (PREC) fuzzy sets
ax0.plot(x_PREC, PREC_VL,'b', linewidth=1.5, label='VL')
ax0.plot(x_PREC, PREC_L ,'g', linewidth=1.5, label='L')
ax0.plot(x_PREC, PREC_N ,'r', linewidth=1.5, label='N')
ax0.plot(x_PREC, PREC_H ,'c', linewidth=1.5, label='H')
ax0.plot(x_PREC, PREC_VH,'m', linewidth=1.5, label='VH')
ax0.plot(x_PREC, PREC_XH,'y', linewidth=1.5, label='XH')
ax0.set_title('Precedentedness (PREC)')
ax0.legend()

# Visualize Process Flexibility (FLEX) fuzzy sets
ax1.plot(x_FLEX, FLEX_VL,'b', linewidth=1.5, label='VL')
ax1.plot(x_FLEX, FLEX_L ,'g', linewidth=1.5, label='L')
ax1.plot(x_FLEX, FLEX_N ,'r', linewidth=1.5, label='N')
ax1.plot(x_FLEX, FLEX_H ,'c', linewidth=1.5, label='H')
ax1.plot(x_FLEX, FLEX_VH,'m', linewidth=1.5, label='VH')
ax1.plot(x_FLEX, FLEX_XH,'y', linewidth=1.5, label='XH')
ax1.set_title('Process Flexibility (FLEX)')
ax1.legend()

# Visualize Risk Resolution (RESL) fuzzy sets
ax2.plot(x_RESL, RESL_VL,'b', linewidth=1.5, label='VL')
ax2.plot(x_RESL, RESL_L ,'g', linewidth=1.5, label='L')
ax2.plot(x_RESL, RESL_N ,'r', linewidth=1.5, label='N')
ax2.plot(x_RESL, RESL_H ,'c', linewidth=1.5, label='H')
ax2.plot(x_RESL, RESL_VH,'m', linewidth=1.5, label='VH')
ax2.plot(x_RESL, RESL_XH,'y', linewidth=1.5, label='XH')
ax2.set_title('Risk Resolution (RESL)')
ax2.legend()


# Visualize Team Cohesion (TEAM) fuzzy sets
ax3.plot(x_TEAM, TEAM_VL,'b', linewidth=1.5, label='VL')
ax3.plot(x_TEAM, TEAM_L ,'g', linewidth=1.5, label='L')
ax3.plot(x_TEAM, TEAM_N ,'r', linewidth=1.5, label='N')
ax3.plot(x_TEAM, TEAM_H ,'c', linewidth=1.5, label='H')
ax3.plot(x_TEAM, TEAM_VH,'m', linewidth=1.5, label='VH')
ax3.plot(x_TEAM, TEAM_XH,'y', linewidth=1.5, label='XH')
ax3.set_title('Team Cohesion (TEAM)')
ax3.legend()

# Visualize Process Maturity (PMAT) fuzzy sets
ax4.plot(x_PMAT, PMAT_VL,'b', linewidth=1.5, label='VL')
ax4.plot(x_PMAT, PMAT_L ,'g', linewidth=1.5, label='L')
ax4.plot(x_PMAT, PMAT_N ,'r', linewidth=1.5, label='N')
ax4.plot(x_PMAT, PMAT_H ,'c', linewidth=1.5, label='H')
ax4.plot(x_PMAT, PMAT_VH,'m', linewidth=1.5, label='VH')
ax4.plot(x_PMAT, PMAT_XH,'y', linewidth=1.5, label='XH')
ax4.set_title('Process Maturity (PMAT)')
ax4.legend()


# Turn off top/right axes
for ax in (ax0, ax1, ax2 ,ax3 ,ax4 ):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()






```

```python

#Find the degree of membership Precedentedness (PREC)
# The exact values 4.99 do not exist on our universes...
# This is what fuzz.interp_membership exists for!
#   *https://pythonhosted.org/scikit-fuzzy/api/skfuzzy.html#interp-membership

PREC_level_VL = fuzz.interp_membership(x_PREC, PREC_VL, 4.5)
PREC_level_L  = fuzz.interp_membership(x_PREC, PREC_L,  4.5)
PREC_level_N  = fuzz.interp_membership(x_PREC, PREC_N,  4.5)
PREC_level_H  = fuzz.interp_membership(x_PREC, PREC_H,  4.5)
PREC_level_VH = fuzz.interp_membership(x_PREC, PREC_VH, 4.5)
PREC_level_XH = fuzz.interp_membership(x_PREC, PREC_XH, 4.5)



print ( 'PREC_level_VL: ',PREC_level_VL )
print ( 'PREC_level_L: ',PREC_level_L  )
print ( 'PREC_level_N: ',PREC_level_N  )
print ( 'PREC_level_H: ',PREC_level_H  )
print ( 'PREC_level_VH: ',PREC_level_VH )
print ( 'PREC_level_XH: ',PREC_level_XH )

# Now we apply this by clipping the top off the corresponding output
# membership function with `np.fmin`
#   https://numpy.org/doc/stable/reference/generated/numpy.fmin.html?highlight=fmin
activation_PREC_VL = np.fmin(PREC_VL,PREC_level_VL)
activation_PREC_L  = np.fmin(PREC_L,PREC_level_L)
activation_PREC_N  = np.fmin(PREC_N,PREC_level_N)
activation_PREC_H  = np.fmin(PREC_H,PREC_level_H)
activation_PREC_VH = np.fmin(PREC_VH,PREC_level_VH)
activation_PREC_XH = np.fmin(PREC_XH,PREC_level_XH)

#print (activation_PREC_VL)


x_PREC0 = np.zeros_like(x_PREC)

# Visualize this degree of membership over Precedentedness (PREC) fuzzy sets
fig, ax5 = plt.subplots(figsize=(8, 3))

# Visualize this Precedentedness Output (PREC)
ax5.fill_between(x_PREC, x_PREC0, activation_PREC_VL, facecolor='b', alpha=0.7)
ax5.plot(x_PREC, PREC_VL,'b', linewidth=0.5, linestyle='--', label='VL')
ax5.fill_between(x_PREC, x_PREC0, activation_PREC_L, facecolor='g', alpha=0.7)
ax5.plot(x_PREC, PREC_L ,'g', linewidth=0.5, linestyle='--', label='L')
ax5.fill_between(x_PREC, x_PREC0, activation_PREC_N, facecolor='r', alpha=0.7)
ax5.plot(x_PREC, PREC_N ,'r', linewidth=0.5, linestyle='--', label='N')
ax5.fill_between(x_PREC, x_PREC0, activation_PREC_H, facecolor='c', alpha=0.7)
ax5.plot(x_PREC, PREC_H ,'c', linewidth=0.5, linestyle='--', label='H')
ax5.fill_between(x_PREC, x_PREC0, activation_PREC_VH, facecolor='m', alpha=0.7)
ax5.plot(x_PREC, PREC_VH,'m', linewidth=0.5, linestyle='--', label='VH')
ax5.fill_between(x_PREC, x_PREC0, activation_PREC_XH, facecolor='y', alpha=0.7)
ax5.plot(x_PREC, PREC_XH,'y', linewidth=0.5, linestyle='--', label='XH')
ax5.set_title('Precedentedness Output (PREC)')
ax5.legend()

```

```python
# Aggregate all  membership functions together
#    https://numpy.org/doc/stable/reference/generated/numpy.fmax.html
aggregated_PREC = np.fmax(activation_PREC_VL, 
                     np.fmax(activation_PREC_L, 
                             np.fmax(activation_PREC_N, 
                                     np.fmax(activation_PREC_H, 
                                             np.fmax(activation_PREC_VH, activation_PREC_XH)))))

# Calculate defuzzified result
PREC = fuzz.defuzz(x_PREC, aggregated_PREC, 'centroid')
activation_PREC = fuzz.interp_membership(x_PREC, aggregated_PREC, PREC)  # for plot

# Visualize this
fig, ax5 = plt.subplots(figsize=(8, 3))

ax5.plot(x_PREC, PREC_VL,'b', linewidth=1.5, label='VL')
ax5.plot(x_PREC, PREC_L ,'g', linewidth=1.5, label='L')
ax5.plot(x_PREC, PREC_N ,'r', linewidth=1.5, label='N')
ax5.plot(x_PREC, PREC_H ,'c', linewidth=1.5, label='H')
ax5.plot(x_PREC, PREC_VH,'m', linewidth=1.5, label='VH')
ax5.plot(x_PREC, PREC_XH,'y', linewidth=1.5, label='XH')
ax5.fill_between(x_PREC, x_PREC0, aggregated_PREC, facecolor='Orange', alpha=0.7)
ax5.plot([PREC, PREC], [0, activation_PREC], 'k', linewidth=1.5, alpha=0.9)

print ('PREC: ', PREC)
```
