```python
import numpy as np
import matplotlib.pyplot as plt


#IS  = improve order process speed.
#PA  = improve payment Accuracy.
#CMS = Camera-man satisfaction.
#CS  = Customer satisfaction.

# set width of bar
barWidth = 0.15
fig = plt.subplots(figsize =(12, 8))
 
# set valus for the bar
IS = [60, 5]
PA = [50, 10 ]
CMS =[70, 3 ]
CS = [29, 3 ]

# Set position of bar on X axis
br1 = np.arange(len(IS))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]

# Make the plot
plt.bar(br1, IS,  color ='r', width  = barWidth,  edgecolor ='grey', label ='IS')
plt.bar(br2, PA,  color ='g', width = barWidth,  edgecolor ='grey', label ='PA')
plt.bar(br3, CMS, color ='b', width = barWidth,  edgecolor ='grey', label ='CMS')
plt.bar(br4, CS,  color ='k', width = barWidth,  edgecolor ='grey', label ='CS')

# Adding Xticks
plt.xlabel('Goal', fontweight ='bold', fontsize = 15)
plt.ylabel('progress', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IS))],['Before', 'After'])
 
plt.legend()
plt.show()
```
